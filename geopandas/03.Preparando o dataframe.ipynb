```python
%matplotlib inline
import pandas as pd
import geopandas as gpd
```

```python
dados = pd.read_table('dados.txt')
```

```python
dados
```

```python
type(dados)
```

# Lembre-se que as coordenadas são sempre P = (longitude, latitude)

```python
from shapely.geometry import Point
```

```python
x = zip(dados.Longitude, dados.Latitude)
```

```python
x
```

```python
geometry = [Point(x) for x in zip(dados.Longitude, dados.Latitude)]
```

```python
geometry
```

```python
crs = {'proj': 'latlong', 'ellps': 'WGS84', 'datum': 'WGS84', 'no_defs': True}
```

```python
crs
```

```python
# transformando dataframe em um geodataframe

geo_dados = gpd.GeoDataFrame(dados, crs = crs, geometry = geometry)
```

```python
geo_dados
```

```python
type(geo_dados)
```

```python
# salvando geodataframe

import os

dir = 'Mapas/RJ-DATASET'
if not os.path.exists(dir):
    os.makedirs(dir)
```

```python
geo_dados.to_file(dir + '/DATASET.shp')
```

```python
# plotando geodataframe

geo_dados.plot(figsize=(15,8), alpha=0.2)
```
